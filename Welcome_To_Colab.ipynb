```
# prompt: check missing and duplicate values from the data

# Check for missing values
missing_values = data.isnull().sum()
print("Missing values:\n", missing_values)

# Check for duplicate rows
duplicate_rows = data.duplicated().sum()
print("\nNumber of duplicate rows:", duplicate_rows)

# Display duplicate rows (optional)
if duplicate_rows > 0:
    print("\nDuplicate rows:\n", data[data.duplicated()])
```

```
# prompt: clean the data

# Check for and handle missing values
# For numerical features, you can fill missing values with the mean or median.
# For categorical features, you can fill missing values with the mode or a separate category.

for col in data.columns:
    if data[col].dtype == 'object':  # Check for string columns
        data[col] = data[col].fillna(data[col].mode()[0]) # Fill with the most frequent value
    elif data[col].dtype in ['int64', 'float64']:  # Check for numerical columns
        data[col] = data[col].fillna(data[col].mean())  # Fill with the mean value

# Remove duplicate rows
data.drop_duplicates(inplace=True)

# Display the cleaned data
print("\nCleaned data:\n")
print(data.head())
print(data.info())
print(data.describe())

# Re-check for missing values and duplicates after cleaning
missing_values_after_cleaning = data.isnull().sum()
print("\nMissing values after cleaning:\n", missing_values_after_cleaning)

duplicate_rows_after_cleaning = data.duplicated().sum()
print("\nNumber of duplicate rows after cleaning:", duplicate_rows_after_cleaning)
```
